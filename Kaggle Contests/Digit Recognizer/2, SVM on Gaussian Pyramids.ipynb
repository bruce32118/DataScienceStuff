```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.cross_validation
import sklearn.svm
import sklearn.neighbors
import sklearn.grid_search
import skimage.transform
import time
%matplotlib inline
```

Here we make use of the methods proposed by <a href="http://www.eecs.berkeley.edu/Pubs/TechRpts/2009/EECS-2009-159.pdf">Maji, 2009</a>.<br/>
Their claim is that in general, the use of raw pixels as features is not as effective as the use of composite features derived from the pixels using various image processing techniques. We begin by applying a Gaussian pyramid reduction to the images.

```python
page_start = time.time()
train = np.array(pd.read_csv('train.csv',header=0))
test = np.array(pd.read_csv('test.csv',header=0))
X, y = train[:,1:], train[:,0]
small_X, small_y = X[:10], y[:10]
```

# Gaussian Pyramid Reduction on Scaled Raw Pixels

The first method is the use of Gaussian pyramid reduction on the raw pixels, with the pixels normaliled such that each pixel's value is between 0 and 1. Our general workflow will be to deterimine an appropriate downscale factor for the pyramid reduction, normalize the data, apply the pyramid reduction, run a grid search to determine the optimal parameters for an SVM estimator, to train that optimal estimator on the training set, and predict on the test set.<br/>
It took a bit of fiddling to figure out how to properly utilize <code>skimage.transform.pyramid_gaussian()</code> in a vectorized manner.
Given an $n\times784$ training set, each record had to be reshaped into an image, giving us an $n\times28\times28$ set. Any transform should operate on the second and third dimensions of our matrix, leaving the number of records $n$ unchanged. However, passing our $n\times28\times28$ matrix directly into <code>skimage.transform.pyramid_gaussian()</code> returned a matrix where the reductions were performed on the first and second dimensions. For instance, the first-order reduction with a downscale factor of 2 performed the following transformation $f: \left(n\times28\times28\right) \to \left(\frac{n}{2}\times14\times28\right).$ The solution turned out to be to pass the transpose of the data to the function, and then to take the transpose of the return value. In other words, given the Gaussian pyramid reduction function $f$ and our data $X$, we were able to retrieve the reduced data $X'$ with $X' = f\left(X^{T}\right)^{T}$.<br/>
Before performing the reduction en-masse, lets see how it affects the human-readability of the data. We'll randomly select one of each digit and show the original, and reduced x1.5, x2, and x3.

```python
f,ax = plt.subplots(nrows=10,ncols=4,figsize=(10,20))
for i in range(10):
    all_of_this_digit = X[y==i]
    random_selection = all_of_this_digit[np.random.choice(len(all_of_this_digit))].astype(float).reshape(28,28)
    ax[i,0].imshow(random_selection,cmap=plt.cm.Greys)
    ax[i,1].imshow(tuple(skimage.transform.pyramid_gaussian(random_selection,downscale=1.5))[1],cmap=plt.cm.Greys)
    ax[i,2].imshow(tuple(skimage.transform.pyramid_gaussian(random_selection,downscale=2))[1],cmap=plt.cm.Greys)
    ax[i,3].imshow(tuple(skimage.transform.pyramid_gaussian(random_selection,downscale=3))[1],cmap=plt.cm.Greys)
ax[0,0].set_title("Original")
ax[0,1].set_title("1.5x Reduce")
ax[0,2].set_title("2x Reduce")
ax[0,3].set_title("3x Reduce")
plt.show()
```

3x reduce looks like a bit too much is lost, but 2x reduce looks promising. Our feature set is reduced from $28\times28=784$ down to $14\times14=196$.<br/>
First, we'll scale the data so that it has pixel value between 0 and 1, and then perform 2x Gaussian pyramid reduction on it.

```python
reduction_start = time.time()
X_normed = X.astype(np.float64)/255.0 # Normalize the data
reshaped_X = X_normed.ravel().reshape(
    # Flatten X with ravel(), and reshape into nx28x28
    len(X),              # n
    np.sqrt(X.shape[1]), # 28
    np.sqrt(X.shape[1])  # 28
)
X_reduced2x = tuple(skimage.transform.pyramid_gaussian(reshaped_X.T,downscale=2))[1].T # Perform the Gaussian pyramid reduction.
X_reduced2x = X_reduced2x.ravel().reshape(len(X_reduced2x),X_reduced2x.shape[1]**2) # reshape from Nx14x14 to Nx196
X_reduced2x = np.ascontiguousarray(X_reduced2x,dtype=np.float64)
print("%d records processed in %0.3f seconds" % (len(X),(time.time()-reduction_start)))
```

## SVM Parameter Selection via Grid Search

Since a priori we don't have a good idea of which model parameters will yield the best performance, we'll perform a grid search to determine the pair of SVM parameters C and gamma that result in the best performing model. We will use a subset of the data to do this search in order to save time.

```python
grid_search_start = time.time()
svm_X,svm_y = X_reduced2x[:0.2*len(X_reduced2x)], y[:0.2*len(y)]
svm_model = sklearn.svm.SVC(kernel='rbf',cache_size=2000)
search_params = {
    'C':10.0**np.arange(-2,5),
    'gamma':10.0**np.arange(-5,0)
}

if __name__ == "__main__":
    grid_search = sklearn.grid_search.GridSearchCV(
        estimator=svm_model,
        param_grid = search_params,
        cv = 5,
        n_jobs = -1 # Use all CPUs in parallel.
    )
    grid_search.fit(svm_X,svm_y)
    print("Best model performance:",grid_search.best_score_)
    print("Model trained in %0.3f seconds" % (time.time() - grid_search_start))
```

```python
best_C, best_gamma = grid_search.best_estimator_.C, grid_search.best_estimator_.gamma
print("Best C: %0.3f, Best Gamma: %0.3f" % (best_C,best_gamma))
```

## Model Fitting on Full Training Set

Now that we have determined the SVM parameters resulting in the highest performance, we can set about training a model using those parameters on the whole dataset:

```python
training_start = time.time()
full_model = sklearn.svm.SVC(
    kernel='rbf',
    C=best_C,
    #C=100.0,
    #gamma=0.1,
    gamma=best_gamma,
    cache_size=2000 # cache size in Mb
)
if __name__ == "__main__":
    full_model_scores = sklearn.cross_validation.cross_val_score(
        estimator=full_model,
        X=X_reduced2x,
        y=y,
        cv=10,
        n_jobs=-1
    )
    full_model.fit(X_reduced2x,y) # Actually fit the model. Interestingly, cross_val_score doesn't appear to fit the model.
print("Full Model Accuracy: %0.2f (+/- %0.2f)" % (full_model_scores.mean(), full_model_scores.std() * 2))
print("Training Time: %0.3f seconds" % (time.time()-training_start))    
```

## Test Set Preprocessing

Now we must normalize and perform the Gaussian pyramid reduction on the test set, after which we can feed it into the model for prediction.

```python
reduction_start = time.time()
test_normed = test.astype(np.float64)/255.0 # Normalize the data
reshaped_test = test_normed.ravel().reshape(
    # Flatten test with ravel(), and reshape into nx28x28
    len(test_normed),              # n
    np.sqrt(test_normed.shape[1]), # 28
    np.sqrt(test_normed.shape[1])  # 28
)
test_reduced2x = tuple(skimage.transform.pyramid_gaussian(reshaped_test.T,downscale=2))[1].T # Perform the Gaussian pyramid reduction.
test_reduced2x = test_reduced2x.ravel().reshape(len(test_reduced2x),test_reduced2x.shape[1]**2) # reshape from Nx14x14 to Nx196
test_reduced2x = np.ascontiguousarray(test_reduced2x,dtype=np.float64)
print("%d records processed in %0.3f seconds" % (len(test_reduced2x),(time.time()-reduction_start)))
```

## Test Set Prediction

```python
prediction_start = time.time()
predictions = full_model.predict(test_reduced2x)
predictions = np.transpose(
    np.array([np.arange(1,len(predictions)+1),np.transpose(predictions)])
)
print("Prediction Time: %0.3f seconds" % (time.time() - prediction_start))
```

## Write Predictions to CSV

```python
np.savetxt(
    'gaussian_pyramid2x.csv',predictions.astype(int),delimiter=",",fmt='%d', header="ImageId,Label"
)
```
