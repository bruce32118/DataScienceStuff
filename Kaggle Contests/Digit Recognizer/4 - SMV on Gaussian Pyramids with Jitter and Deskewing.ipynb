# Data Preparation: Deskew, Gaussian Pyramid reduction, and pixel offset jitter.

We will deskew the images using an affine transformation on image moments, reduce the deskewed images to $12\times12$ pixels using Gaussian Pyramids, and then apply a pixel offset jitter to the images. We'll do this by taking each $12\times12$ image and inlaying it in a $14\times14$ area (effectively adding an additional row on the top and bottom, and an additional column on the left and right). There will then be $3\times3=9$ different ways that we can position our $12\times12$ image in this $14\times14$ area, in the same way that there are $9$ different $1\times1$ arrangements in a $3\times3$ matrix. By taking each of the $9$ offset combinations for each image, we'll effectively multiply our sample size by $9$, and hopefully be able to train a more robust classifier in the process.

### Deskew

We deskew by applying an affine transformation to each image, using the <a href="http://en.wikipedia.org/wiki/Image_moment">central</a> moments about the image mean.

```python
def deskew(img):
    M = skimage.measure.moments(img) # Take the raw image moments
    M = M/M[0,0] # Normalize
    mu = skimage.measure.moments_central(img,M[1,0],M[0,1]) # Take the moments about the image mean. M10 and M01 are centroid coords.
    skew = mu[1,1]/mu[0,2]
    transformation_matrix = np.array([
        [1,skew,-0.5*img.shape[0]*skew], # shear in X direction by skew, translate
        [0,1,0],
        [0,0,1]
    ])
    transformation = skimage.transform.AffineTransform(matrix=transformation_matrix)
    return skimage.transform.warp(img,transformation)
```

```python
deskew_start = time.time()
pool = multiprocessing.dummy.Pool() # Process in parallel to save time.
deskewed_X = np.array(pool.map(deskew,X.ravel().reshape(len(X),28,28).astype(np.float)/255.0))
deskewed_test = np.array(pool.map(deskew,test.ravel().reshape(len(test),28,28).astype(np.float)/255.0))
print("Deskewed %d images in %0.3f seconds" % (len(deskewed_X)+len(deskewed_test),(time.time()-deskew_start)))
```

```python
f,ax = plt.subplots(nrows=10,ncols=2,figsize=(5,10))
for i in range(10):
    all_of_this_digit = X[y==i]
    random_selection = all_of_this_digit[np.random.choice(len(all_of_this_digit))]
    random_selection = random_selection.reshape(28,28).astype(np.float)/255.0
    ax[i,0].imshow(random_selection,cmap=plt.cm.Greys)
    ax[i,1].imshow(deskew(random_selection),cmap=plt.cm.Greys)
    ax[i,0].axes.get_xaxis().set_visible(False)
    ax[i,0].axes.get_yaxis().set_visible(False)
    ax[i,1].axes.get_xaxis().set_visible(False)
    ax[i,1].axes.get_yaxis().set_visible(False)
ax[0,0].set_title("Original")
ax[0,1].set_title("Affine")
plt.show()
```

### Gaussian Pyramid Reduction

Now we'll perform Gaussian Pyramid reduction on the deskewed images to reduce them from $28\times28$ to $12\times12$.

```python
pyramid_start = time.time()
# Thank god this function supports vectorized operations...
X_reduced12x12 = tuple(skimage.transform.pyramid_gaussian(deskewed_X.T,downscale=1.35))[3].T
test_reduced12x12 = tuple(skimage.transform.pyramid_gaussian(deskewed_test.T,downscale=1.35))[3].T
print("Reduced in %0.3f seconds" % (time.time()-pyramid_start))
```

### Pixel offset jitter

```python
pixel_offset_start = time.time()
X_jittered = []
y_jittered = []
test_centered = []
for i,img in enumerate(X_reduced12x12):
    for j in range(3):
        for k in range(3):
            new_img = np.zeros((14,14))
            new_img[j:j+12,k:k+12] += img
            X_jittered.append(new_img)
            y_jittered.append(y[i])
for img in test_reduced12x12:
    new_img = np.zeros((14,14))
    new_img[1:13,1:13] += img
    test_centered.append(new_img)
X_jittered = np.ascontiguousarray(np.array(X_jittered).ravel().reshape(len(X_jittered),-1),dtype=np.float)
y_jittered = np.ascontiguousarray(y_jittered,dtype=np.int)
test_centered = np.ascontiguousarray(np.array(test_centered).ravel().reshape(len(test_centered),-1),dtype=np.float)
print("Created %d samples in %0.3f seconds" %(len(X_jittered)+len(test_centered),(time.time()-pixel_offset_start)))
```

From this perspective, the effect of the jitter can be seen: each digit has been moved around the image. The fifth row shows the original centered images.

### Random Image Rotation

We'll test if our model's robustness is improved by randomly rotating our images by slight amounts. Our degrees of rotation will be random samples from a normal distribution with $\mu=0$ and $\sigma=4$, rounded to the nearest integer.

### SVM Training

We'll train an SVM using the now fully-preprocessed image data.

```python
training_start = time.time()
model = sklearn.svm.SVC(kernel='rbf',C=100.0,gamma=0.1,cache_size=2000)
if __name__ == '__main__':
    cross_validation_scores = sklearn.cross_validation.cross_val_score(estimator=model,X=X_jittered,y=y_jittered,cv=10,n_jobs=-1)
    model.fit(X_jittered,y_jittered)
    print("Estimated model accuracy: %0.3f +/- %0.3f" %(cross_validation_scores.mean(),cross_validation_scores.std()))
print("Number of features: %d\nNumber of Samples: %d\nTraining time: %0.3f seconds" %
      (X_jittered.shape[1],X_jittered.shape[0],(time.time()-training_start))
)
```

```python
prediction_start = time.time()
predictions = model.predict(test_centered)
predictions = np.transpose(
    np.array([np.arange(1,len(predictions)+1),np.transpose(predictions)])
)
print("Prediction Time: %0.3f seconds" % (time.time() - prediction_start))
```
