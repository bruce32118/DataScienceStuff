Here we'll attempt to develop a method of automatically deskewing the images through the use of <a href="http://en.wikipedia.org/wiki/Image_moment">image moments</a>. The general method is to take an image, compute its second-order central moments (which correspond to variance), and use these to derive a covariance matrix, the eigenvectors of which tell us the image's major and minor axes, and thereby the image's orientation. The angular difference between these axes and the $x$-$y$ plane, derived using the dot product, will allow us to then deskew the image.

```python
import numpy as np
import pandas as pd
import skimage.transform
import skimage.measure
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = np.array(pd.read_csv('train.csv',header=0))
test = np.array(pd.read_csv('test.csv',header=0))
X, y = train[:,1:], train[:,0]
small_X, small_y = X[:10], y[:10]
```

```python
sample_images = X[[0,3,516]].ravel().reshape(3,28,28).astype(np.float64)/255.0
f,ax = plt.subplots(nrows=1,ncols=3)
ax[0].imshow(sample_images[0],cmap=plt.cm.Greys)
ax[1].imshow(sample_images[1],cmap=plt.cm.Greys)
ax[2].imshow(sample_images[2],cmap=plt.cm.Greys)
plt.show()
```

# Deskewing via Rotation of Major Axis of Variance

```python
def deskew(img):
    M = skimage.measure.moments(img)
    M = M/M[0,0]
    mu = skimage.measure.moments_central(img,M[1,0],M[0,1])
    #mu = mu/mu[0,0]
    cov = np.array([[mu[2,0],mu[1,1]],[mu[1,1],mu[0,2]]])
    eigvals,eigvecs = np.linalg.eig(cov)
    major_axis = eigvecs[eigvals.argmax()]
    d = (np.pi - np.arccos(np.dot(major_axis,[1,0])))*180.0/np.pi
    return skimage.transform.rotate(img,angle=90.0-d)
```

```python
f,ax = plt.subplots(nrows=len(sample_images),ncols=2)
for i,img in enumerate(sample_images):
    ax[i,0].imshow(img,cmap=plt.cm.Greys)
    ax[i,1].imshow(deskew(img),cmap=plt.cm.Greys)
plt.show()
```

This gives us mixed results. Some get deskewed correctly, whereas others don't (for example, many 4's get turned on their sides at odd angles).

# Affine Transformations

```python
def deskew_affine(img):
    M = skimage.measure.moments(img)
    M = M/M[0,0]
    mu = skimage.measure.moments_central(img,M[1,0],M[0,1])
    mu = mu/mu[0,0]
    skew = mu[1,1]/mu[0,2]
    transformation_matrix = np.array([
        [1,skew,-0.5*img.shape[0]*skew], # shear in X direction by skew, translate
        [0,1,0],
        [0,0,1]
    ])
    transformation = skimage.transform.AffineTransform(matrix=transformation_matrix)
    return skimage.transform.warp(img,transformation)

plt.imshow(deskew_affine(sample_images[0]),cmap=plt.cm.Greys)
plt.show()
```

```python
f,ax = plt.subplots(nrows=len(sample_images),ncols=3)
for i,img in enumerate(sample_images):
    ax[i,0].imshow(img,cmap=plt.cm.Greys)
    ax[i,1].imshow(deskew(img),cmap=plt.cm.Greys)
    ax[i,2].imshow(deskew_affine(img),cmap=plt.cm.Greys)
ax[0,0].set_title("Original")
ax[0,1].set_title("Eigenvector")
ax[0,2].set_title("Affine")
plt.show()
```

```python
f,ax = plt.subplots(nrows=10,ncols=3,figsize=(10,20))
for i in range(10):
    all_of_this_digit = X[y==i]
    random_selection = all_of_this_digit[np.random.choice(len(all_of_this_digit))]
    random_selection = random_selection.reshape(28,28).astype(np.float)/255.0
    ax[i,0].imshow(random_selection,cmap=plt.cm.Greys)
    ax[i,1].imshow(deskew(random_selection),cmap=plt.cm.Greys)
    ax[i,2].imshow(deskew_affine(random_selection),cmap=plt.cm.Greys)    
ax[0,0].set_title("Original")
ax[0,1].set_title("Eigenvector")
ax[0,2].set_title("Affine")
plt.show()
```
