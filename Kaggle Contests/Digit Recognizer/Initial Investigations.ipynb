```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimage
import scipy.misc
import scipy.ndimage as ndimage
import skimage.filter as skif
import sklearn
from sklearn import neighbors, linear_model
%matplotlib inline
```

First order of business is to see what the images look like. 
Initially we'll use two methods of image processing. The first is scikit-image's adaptive threshold filter,
and the second involves decomposing the image matrix using the SVD (singular value decomosition), zeroing a number
of the lesser singular values, and recomposing the matrix into an image. This gives us a "low-dimensional" representation
of our image, to see if it's form is any clearer. We'll also combined adaptive thresholding with SVD reduction and see how that works.

```python
# Get the data into a dataframe.
train = pd.read_csv('train.csv',header=0)
test = pd.read_csv('test.csv',header=0)

small_train = train.iloc[:100,:]

f, axarr = plt.subplots(10, 4,figsize=(9,9))
axarr[0, 0].set_title('Original')
axarr[0, 1].set_title('Adaptive Thresholding')
axarr[0, 2].set_title('SVD Reduction')
axarr[0, 3].set_title('SVD + Adaptive')

for i in range(10):
    sample = small_train[small_train['label'] == i]
    test_row = sample.iloc[np.random.random_integers(0,len(sample)-1),:]
    
    # Get test image:
    (label,img) = (test_row[0],test_row[1:].reshape(28,28))

    # Adaptive thresholding
    adaptive_img = skif.threshold_adaptive(img,block_size=50,offset=0)

    # Dimensionality Reduction via SVD:
    singular_values = 8 # The number of singular values to keep.
    (U,s,V) = np.linalg.svd(img)
    s[singular_values:] = 0
    svd_img = np.dot(U,np.dot(np.diag(s),V))

    # Apply adaptive thresholding to SVD-reduced image:
    svd_adaptive_img = skif.threshold_adaptive(svd_img,block_size=50,offset=0)
    
    # Plot the raw image, as well as with adaptive thresholding.
    axarr[i, 0].imshow(img,cmap=plt.cm.Greys)
    axarr[i, 1].imshow(adaptive_img,cmap=plt.cm.Greys)
    axarr[i, 2].imshow(svd_img,cmap=plt.cm.Greys)
    axarr[i, 3].imshow(svd_adaptive_img,cmap=plt.cm.Greys)
    for j in range(4):
        axarr[i,j].get_yaxis().set_visible(False)
        axarr[i,j].get_xaxis().set_visible(False)
```

It doesn't appear that SVD does much good. Adaptive thresholding, however shows significant promise.

## Image Downsizing

### Pyramid Reduction

```python
import skimage.transform
img = train.iloc[7,1:].reshape(28,28)
img1 = skimage.transform.pyramid_reduce(
    img.astype(np.float64),downscale=1.5
)
img2 = skimage.transform.pyramid_reduce(
    img.astype(np.float64),downscale=2
)
f, ax = plt.subplots(nrows=1,ncols=3)
ax[0].imshow(img)
ax[0].set_title("Full-size Image")
ax[1].imshow(img1)
ax[1].set_title("Reduced x1.5")
ax[2].imshow(img2)
ax[2].set_title("Reduced by x2")
plt.show()
```

Gets rid of too much. 3 looks like an 8.

### Block Reduction

```python
img1 = skimage.measure.block_reduce(img,block_size=(2.0,2.0))
img2 = skimage.measure.block_reduce(img,block_size=(3.0,3.0))
f, ax = plt.subplots(nrows=1,ncols=3)
ax[0].imshow(img)
ax[0].set_title("Full-size Image")
ax[1].imshow(img1)
ax[1].set_title("Reduced x2")
ax[2].imshow(img2)
ax[2].set_title("Reduced by x3")
plt.show()
```

Bingo. There we go.
